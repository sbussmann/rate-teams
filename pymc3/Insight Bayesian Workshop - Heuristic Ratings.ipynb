```python
import os
import pandas as pd
```

```python
base_dir = os.path.join("/Users", "sbussmann", "Development", "buda", "buda-ratings")
```

```python
interim_dir = os.path.join(base_dir, 'data', 'interim')
```

```python
league_id = 40264
```

```python
plusminus = pd.read_csv(os.path.join(interim_dir, 'plusminus_{}.csv'.format(league_id)))
```

```python
plusminus.head()
```

```python
winpct = plusminus['Wins'] / (plusminus['Wins'] + plusminus['Losses'])
```

```python
plusminus = plusminus.assign(winpercent = winpct)
```

```python
plusminus.sort_values('PlusMinus', ascending=False)
```
